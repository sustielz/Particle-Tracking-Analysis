An LED is illuminated briefly at the zero-crossing of a sine wave. If we record the LED with a camera, how can we measure the phase of the zero-crossing?

Let the camera have framerate $F$, and let the LED flash for a duration $dt$ at frequency $f$, choosing $dt < 1/F$ so that a single flash does not illuminate the LED for more than one frame. Further, suppose that we record a video of $N$ frames, or duration $T=N/FPS$. 




If we detect a flash on frame $n$, then we know that the LED must have flashed within the interval $\tau_n^0 \in [n/F - dt, n/F]$. Furthermore, the LED will ONLY flash some integer $i$ flashes later (or earlier, if $i<0$) at time $\tau_n^i = \tau_n^0 + i/f \in [n/F + i/f - dt, n/F + i/f]$; Likewise, if a flash is NOT detected at frame $n$, then NO flashes can occur in any interval $[n/F + i/f - dt, n/F + i/f]$.




Next, suppose another flash is detected some $\Delta N$ frames after $t=0$ (or before $t=0$, if $\Delta N$ negative). Because the LED is periodic, this flash must have occured some integer $n$ frames after (before) $t=0$, so we have $\phi + n/f \in [(\Delta N) T - dt, (\Delta N) T]$, or $\phi \in [\Delta N/F - n/f - dt, \Delta N/F - n/f]$.  


Note that we also have $\phi \in [-dt, 0]$, so the $n$'th frame is detectable only when $\Delta N \in [nF/f, nF/f + Fdt]$. Furthermore, since we have chosen $dt < T$, the length of this interval is $Fdt<1$ so each flash $n$ is detectable by at most one frame $\Delta N$. If this interval contains no integers, then the $n$'th flash is not detectable at all. 

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams['figure.dpi'] = 200
import cv2, json

f = 6.18
dt = 20/1000
FPS = 30

def load_intensity(filename, path='', thresh=0.5):  ## Load data from json file
    filename = filename.split('.')[0]  ## Precaution, in case filename includes suffix
    if path is not '' and path[-1] is not '/':
        path = path + '/'
    with open(path+'Intensity/'+filename+'.json', 'r') as f:
        preds =  json.load(f)
    thresh = min(preds) + thresh*( max(preds) - min(preds) )
    out = np.zeros(np.size(preds))
    out[np.array(preds)>thresh] += 1
    return out 

def intersect(i, j):
    dT = abs(j - i)/FPS
    high = 0
    while dT > high:
        high += 1/f
        
    if high - dT < dt:
        return [i/FPS - dt, i/FPS - (high - dT)]
    elif dT - (high-1/f) < dt:
        return [i/FPS + dT - (high-1/f) - dt, i/FPS]    
    
```

```python
I = load_intensity('6pt18hz_vpp10_long1.json', path='feb12')
plt.plot(I)
```

```python
# N = np.size(I)
N = 100
for i in range(N):
    if I[i] > 0.5:
        plt.plot([i/FPS - dt, i/FPS], [i+1, i+1], color='green')
        
        j=1
        while(i/FPS + j/f < N/FPS):
            plt.plot([i/FPS + j/f - dt, i/FPS + j/f], [i+1, i+1], color='blue')
            j += 1
        j=-1
        while(i/FPS + j/f > 0):
            plt.plot([i/FPS + j/f - dt, i/FPS + j/f], [i+1, i+1], color='blue')
            j -= 1
    
```

```python




N = np.size(I)
# N = 100
flashes = []
nonflashes = []
for i in range(N):
    if I[i] > 0.5:
        flashes.append(i)
    else:
        nonflashes.append(i)

i0 = flashes[0]
ivls = []
lows = []
highs = []

nivls = []
nlows = []
nhighs = []

plt.plot([i0/FPS - dt, i0/FPS], [1, 1], color='green')
for i in range(np.size(flashes)):
    ivl = intersect(flashes[0], flashes[i])
    lows.append(ivl[0])
    highs.append(ivl[1])
    ivls.append(ivl)
    
    plt.plot(ivl, [i+1, i+1], color='blue')
    
plt.show()
for i in range(np.size(nonflashes)):
    ivl = intersect(nonflashes[0], nonflashes[i])
    if ivl is not None:
        nlows.append(ivl[0])
        nhighs.append(ivl[1])
        nivls.append(ivl)
    
        plt.plot(ivl, [i+1, i+1], color='red')
       
```
